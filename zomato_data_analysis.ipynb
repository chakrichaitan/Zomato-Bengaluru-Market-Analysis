```python
import pandas as pd
import numpy as np

class RestaurantAnalyzer:
    """
    A class to load, clean, and analyze the Zomato restaurant data.
    """
    def __init__(self, filepath):
        """
        Constructor for the analyzer.
        Initializes with the path to the dataset.
        """
        self.filepath = filepath
        self.df = None  # We initialize the dataframe as None. It will be loaded by our method.
        print("RestaurantAnalyzer object created.")

    def load_data(self):
        """
        Loads the data from the CSV file into a pandas DataFrame.
        Includes exception handling for file errors.
        """
        try:
            self.df = pd.read_csv(self.filepath)
            print(f"Data loaded successfully. Shape: {self.df.shape}")
        except FileNotFoundError:
            print(f"Error: The file at {self.filepath} was not found.")
            print("Please check the file path.")
            self.df = None # Ensure df is None if file is not found
        
        return self.df
```

# Step 3: Data Wrangling & Cleaning

#### 1. Create a Working Copy

```python
# Create a copy to keep the original data safe
df_cleaned = df.copy()
```

#### 2. Dropping Unnecessary Columns

```python
# Drop the columns that are not required for our analysis
df_cleaned = df_cleaned.drop(columns=['url', 'dish_liked', 'phone'])

# Let's check the result
print("Columns after dropping:")
print(df_cleaned.columns)
print(f"\nNew shape of the data: {df_cleaned.shape}")
```

#### 3. Cleaning the Column Names

```python

# Rename columns for easier access
df_cleaned.rename(columns={
    'approx_cost(for two people)': 'approx_cost',
    'listed_in(type)': 'meal_type',
    'listed_in(city)': 'area'
}, inplace=True)

# Check the new column names
print("Renamed columns:")
print(df_cleaned.columns)
```

#### 4. rate Column

```python
# First, let's see all the unique values to understand the problem
print("Unique values in 'rate' column before cleaning:")
print(df_cleaned['rate'].unique())
```

```python
# Let's define a rule (as a small, one-line function) to clean these values
# We use a 'lambda' function here, which is just a way to write a quick, anonymous function
def clean_rate(rate_str):
    if isinstance(rate_str, str):
        if "NEW" in rate_str or "-" in rate_str:
            return np.nan  # Return a missing value for 'NEW' and '-'
        else:
            return float(rate_str.split('/')[0]) # Split "4.1/5" and take the "4.1" part
    return rate_str # Keep it as is if it's already a number or NaN
```

```python
# Apply this rule to the entire 'rate' column
df_cleaned['rate'] = df_cleaned['rate'].apply(clean_rate)
```

```python
print("\nUnique values in 'rate' column after cleaning:")
print(df_cleaned['rate'].unique())
print(f"\nData type of 'rate' column now: {df_cleaned['rate'].dtype}")
```

# WHAT WE DID TILL NOW
We first investigated the unique values to see all the different kinds of "dirt" in the data.

We defined a cleaning rule in a function clean_rate which intelligently handles each case.

We used the powerful .apply() method to run our cleaning rule on every single value in the rate column.

The column is now of type float64 and contains only numbers or NaN (Not a Number), which represents the missing values.

#### 5. Cleaning the approx_cost Column

```python
<> We need to remove the comma from strings like "1,500" and convert them to numbers.
```

```python
# Check the unique values first
print("Unique values in 'approx_cost' column before cleaning:")
print(df_cleaned['approx_cost'].unique()[:20]) # Show first 20 unique values
```

```python
# Remove commas and convert to numeric
df_cleaned['approx_cost'] = df_cleaned['approx_cost'].str.replace(',', '', regex=False)
df_cleaned['approx_cost'] = pd.to_numeric(df_cleaned['approx_cost'])
```

```python
print("\nUnique values in 'approx_cost' column after cleaning:")
print(df_cleaned['approx_cost'].unique()[:20])
print(f"\nData type of 'approx_cost' column now: {df_cleaned['approx_cost'].dtype}")
```

```python
df_cleaned.info()
```

#### 6. Handling the Remaining Missing Values

```python
Our data types are now correct. It's finally time to fill in the gaps
```

```python
# For numerical columns like 'rate' and 'approx_cost', filling with the median is a good strategy
# because it is not affected by extreme outliers.
rate_median = df_cleaned['rate'].median()
df_cleaned['rate'].fillna(rate_median, inplace=True)

cost_median = df_cleaned['approx_cost'].median()
df_cleaned['approx_cost'].fillna(cost_median, inplace=True)
```

```python
number>>>int, float>>>mean or median>>>outliers median>>>no outliers mean
categorical columns>>> mode
```

```python
# For categorical columns, filling with the mode (most frequent value) is the best approach.
location_mode = df_cleaned['location'].mode()[0]
df_cleaned['location'].fillna(location_mode, inplace=True)

rest_type_mode = df_cleaned['rest_type'].mode()[0]
df_cleaned['rest_type'].fillna(rest_type_mode, inplace=True)

cuisines_mode = df_cleaned['cuisines'].mode()[0]
df_cleaned['cuisines'].fillna(cuisines_mode, inplace=True)

print("Missing values have been handled.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for all our plots
sns.set_style('darkgrid')
# Set the default figure size
plt.rcParams['figure.figsize'] = (10, 7)

print("Plotting libraries imported and styles set.")
```

### What is the distribution of restaurant ratings? A histogram is the perfect plot for this.

```python
# Create a histogram of the 'rate' column
sns.histplot(df_cleaned['rate'], bins=20, kde=True)
plt.title('Distribution of Restaurant Ratings', fontsize=14)
plt.xlabel('Rating (out of 5)', fontsize=14)
plt.ylabel('Number of Restaurants', fontsize=14)
plt.show()
```

**Analysis of "Distribution of Restaurant Ratings":
    
The Crowd is in the Middle: The first thing we see is a huge peak in the distribution between ratings of 3.5 and 4.2. This tells us that the vast majority of restaurants in Bengaluru fall into the "average" to "good" category.

It's a Tough Market: The distribution is left-skewed. This means the tail on the left side (lower ratings) is longer than on the right. In simple terms, it's easier to be a mediocre restaurant (rating 3.0-3.5) than it is to be an excellent one (rating 4.5+). The bar for excellence is very high.

Key Insight: For our entrepreneur client, this means that just opening a "good" restaurant isn't enough to stand out. The market is saturated with them. The real goal should be to break into the elite 4.3+ category. The median rating we used to fill missing values was likely in that 3.7-3.8 range, which seems like a very reasonable choice now.

**Is there a difference in ratings between restaurants that offer online ordering and those that don't? 

#### Do restaurants that offer online ordering have better ratings? box plot.

```python
# Let's compare the ratings of restaurants with and without online ordering
sns.boxplot(x='online_order', y='rate', data=df_cleaned)
plt.title('Rating vs. Online Order Availability', fontsize=14)
plt.xlabel('Offers Online Order?', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.show()
```

The Median Story: The median rating (the line in the middle of the box) for restaurants with online ordering is around 3.7, while for those without, it's closer to 3.4.

Key Insight: This provides powerful evidence for our Hypothesis #3 ("The Location & Convenience Factor"). Restaurants that embrace the convenience of online ordering tend to have significantly higher ratings. This could be for two reasons: 1) Better, more popular restaurants are more likely to offer online services, or 2) The convenience itself improves customer satisfaction and ratings. Either way, for our client, offering online ordering is non-negotiable.

#### What is the relationship between cost and rating?

This directly tests our Hypothesis 1 (The "Price vs. Quality" Myth). We want to see if more expensive restaurants actually have better ratings.

For two numerical columns (approx_cost and rate), we'd normally use a scatter plot. However, with over 50,000 data points, a regular scatter plot would be a messy, unreadable blob. A much more professional choice is a joint plot with a hexagonal binning (kind='hex'), which shows the density of points.

```python
# Analyze the relationship between approximate cost and rating
# A jointplot with hexbin is great for visualizing dense data
sns.jointplot(x='approx_cost', y='rate', data=df_cleaned, kind='hex', cmap='viridis_r')
plt.suptitle('Density of Cost vs. Rating', y=1.02, fontsize=14 ) # Add a main title
plt.show()
```

Analysis of "Density of Cost vs. Rating" (Joint Plot)
The Sweet Spot: The darkest area of the plot (where the highest density of restaurants is) is in the low-cost region (approx_cost under â‚¹1,000) and the high-rating region (rate between 3.6 and 4.2).

High Cost â‰  High Rating: As we move to the right along the x-axis (increasing cost), the ratings do not consistently increase. In fact, there are many very expensive restaurants (cost > â‚¹2,000) with mediocre ratings (3.0-3.5).

Key Insight: This plot is a fantastic confirmation of our Hypothesis #1 ("The Price vs. Quality" Myth). You do not need to open an expensive, fine-dining restaurant to be successful in Bengaluru. The most successful segment of the market is the one that offers high-quality food at an affordable price. This is a game-changing insight for our entrepreneur client's business plan.

#### Does offering table booking impact ratings?

```python
# Analyze the impact of table booking on ratings
sns.boxplot(x='book_table', y='rate', data=df_cleaned)
plt.title('Rating vs. Table Booking Availability', fontsize=14)
plt.xlabel('Offers Table Booking?', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.show()
```

A Massive Difference: This is perhaps the strongest signal we've seen yet. Restaurants that offer table booking have a median rating of around 4.2, which is exceptionally high. Those that don't have a median rating of around 3.7.

A Different Class of Restaurant: This tells us that the "book table" feature is a clear marker for premium, dine-in focused establishments. These are places people plan to go to, rather than just dropping in.

Key Insight : While online ordering is crucial for the mass market, offering table booking is a hallmark of the top tier of restaurants. For a client wanting to open a high-quality establishment, this feature is essential.

#### Which restaurant types are the most highly rated?

```python
# To make the plot readable, let's increase the figure size
plt.figure(figsize=(12, 6))
# Analyze ratings across different restaurant types
sns.boxplot(x='rest_type', y='rate', data=df_cleaned)
plt.title('Rating vs. Restaurant Type', fontsize=14)
plt.xlabel('Restaurant Type', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=90) # Rotate the x-axis labels to prevent overlap
plt.show()
```

"Experience" Dining Wins: Although the plot is crowded, we can see a clear trend by looking at the top performers. Restaurant types like Breweries, Pubs, Lounges, and Fine Dining consistently have higher median ratings.

"Functional" Dining is Lower: On the other end, types like Food Courts, Quick Bites, and many Delivery-only places have lower median ratings.

Key Insight: People rate the entire experience. Venues that offer a great ambiance, service, and a social setting (like breweries and pubs) are rewarded with higher ratings than places that simply provide a quick meal.

#### Which areas have the highest-rated restaurants?

>> area column has too many unique values to plot directly. 
>> identify the most important areas (e.g., the top 15 areas with the most restaurants) and then analyze the ratings within that group. A bar plot showing the average rating is best here.

```python
# Step 1: Find the top 15 areas with the most restaurants
top_areas = df_cleaned['area'].value_counts().nlargest(15).index
```

```python
# Step 2: Filter the DataFrame to include only these top areas
df_top_areas = df_cleaned[df_cleaned['area'].isin(top_areas)]
```

```python

# Step 3: Create the plot
plt.figure(figsize=(15, 6))
sns.boxplot(x='area', y='rate', data=df_top_areas, palette='viridis')
plt.title('Ratings in Top 15 Areas', fontsize=14)
plt.xlabel('Area', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

Consistently High Standards: The first thing to note is that all top 15 areas are highly competitive, with median ratings mostly above 3.8. There are no "bad" areas in this top list.

The Premier Hubs: However, we can see a clear elite group. Areas like Brigade Road, Church Street, MG Road, and Residency Road show the highest median ratings, often centered around 4.1-4.2. These are the premium dining hotspots in Bengaluru.

Key Insight : Location is critical. While being in any of these top 15 areas is good, aiming for the central, premium hubs like Church Street or Brigade Road puts a new restaurant in the best possible neighborhood for attracting high ratings.

#### The Final Piece of Analysis: What Cuisines are a Hit?

```python
# Find the top 10 most common cuisines
top_cuisines = df_cleaned['cuisines'].value_counts().nlargest(10).index

# Filter the DataFrame to include only these top cuisines
df_top_cuisines = df_cleaned[df_cleaned['cuisines'].isin(top_cuisines)]

# Let's plot their ratings
plt.figure(figsize=(15, 6))
sns.boxplot(x='cuisines', y='rate', data=df_top_cuisines, palette='inferno')
plt.title('Ratings of Top 10 Most Popular Cuisines', fontsize=14)
plt.xlabel('Cuisine', fontsize=14)
plt.ylabel('Rating (out of 5)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

"Safe" but Not "Spectacular": The most common cuisines like North Indian and South Indian have very standard rating distributions, with medians around 3.7. They are incredibly popular but also highly competitive, making it hard to stand out.

The "Jack of all Trades" Trap: Look at the cuisines on the far right: South Indian, North Indian, Chinese. The median rating for these multi-cuisine restaurants is visibly lower than for more specialized ones. This is strong evidence for our Hypothesis #2, confirming that restaurants that try to do everything are often perceived as masters of none.

The High Performers: Among the most popular options, Desserts and Cafe stand out with a higher median rating and a strong cluster of highly-rated establishments. This suggests that a well-executed, specialized concept in these areas is a recipe for success.

Key Insight : Specialization is key. Instead of offering a generic mix of popular cuisines, a new restaurant would be better served by excelling in a specific, high-performing category like Desserts or Cafe, or by offering a premium version of a standard cuisine.
